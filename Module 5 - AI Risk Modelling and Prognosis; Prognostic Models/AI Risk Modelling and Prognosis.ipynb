## Simulating NRI Calculation
NRI measures whether a new model correctly reclassifies patients compared to a 
baseline. Below is a simplified simulation across a small patient sample.

```python
patients = {
    "Patient": [1, 2, 3, 4, 5, 6],
    "Has Disease": [1, 1, 1, 0, 0, 0],
    "Baseline Risk": [0.4, 0.3, 0.6, 0.5, 0.2, 0.4],
    "New Model Risk": [0.7, 0.6, 0.8, 0.2, 0.1, 0.3]
}

df = pd.DataFrame(patients)

df["Reclassified Up"] = df["New Model Risk"] > df["Baseline Risk"]
df["Reclassified Down"] = df["New Model Risk"] < df["Baseline Risk"]

nri_events = df[df["Has Disease"] == 1]["Reclassified Up"].sum() - \
             df[df["Has Disease"] == 1]["Reclassified Down"].sum()

nri_non_events = df[df["Has Disease"] == 0]["Reclassified Down"].sum() - \
                 df[df["Has Disease"] == 0]["Reclassified Up"].sum()

nri_total = (nri_events + nri_non_events) / len(df)

print(f"NRI Events: {nri_events}")
print(f"NRI Non-Events: {nri_non_events}")
print(f"Total NRI Score: {nri_total:.2f}")
df
```

## Takeaway
NRI is a useful metric but requires pairing with additional evaluation methods 
to rule out chance-driven reclassification. The validation gap in LMIC settings 
is not a minor oversight. It represents a structural failure that directly affects 
patient outcomes in the environments that carry the highest disease burden.
