## Intervention Decision Logic
A simplified representation of how the agent selects actions based on mood score.

```python
def select_intervention(mood_score):
    if mood_score >= 7:
        return "Do nothing"
    elif mood_score >= 5:
        return "Deliver breathing exercise"
    elif mood_score >= 3:
        return "Prompt journaling session"
    else:
        return "Escalate to human professional"

sample_states = pd.DataFrame({
    "User": ["User A", "User B", "User C", "User D"],
    "Mood Score (0-10)": [8, 5, 3, 1]
})

sample_states["Recommended Action"] = sample_states["Mood Score (0-10)"].apply(select_intervention)
sample_states
```

## Takeaway
The agent's value depends entirely on how responsibly it is governed. Timing 
and personalization are just as consequential as the intervention itself, and 
human escalation must be a meaningful pathway, not a last resort.
