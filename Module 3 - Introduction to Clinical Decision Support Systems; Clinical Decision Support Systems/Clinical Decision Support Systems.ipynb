## Simulating the Risk Score
The CDSS in the assignment calculated a 72% probability of deterioration within 
12 hours by combining HRV analysis with incoming lab values. Below is a simplified 
representation of how weighted clinical inputs produce a composite risk score.

```python
patient_data = {
    "Feature": ["Heart Rate Variability", "Lactate", "Creatinine", "INR", "Heart Rate"],
    "Value": [0.38, 2.6, 1.5, 1.4, 96],
    "Weight": [0.30, 0.25, 0.20, 0.15, 0.10]
}

df = pd.DataFrame(patient_data)

max_values = [1.0, 4.0, 2.0, 2.0, 120]
df["Normalized"] = [v / m for v, m in zip(df["Value"], max_values)]
df["Weighted Score"] = df["Normalized"] * df["Weight"]

risk_score = df["Weighted Score"].sum()
print(f"Composite Risk Score: {risk_score:.2f}")
print(f"Deterioration Probability: {risk_score * 100:.1f}%")
df
```

## Takeaway
Real CDSS platforms use far more complex algorithms, but the core logic is the same: 
clinical inputs are weighted, normalized, and combined into a score that supports 
rather than replaces physician judgment.
