```python
import pandas as pd
```

```python
## Load the dataset 
df = pd.read_csv("healthcare_dataset.csv")

## Preview first rows to confirm it loaded correctly
df.head(2)
```

```python
## Calculate the average age of all people in the dataset
average_age = df["Age"].mean()
average_age
```

```python
## Calculate the average resting heart rate of all people in the dataset
average_heartbeat = df["Heartbeat"].mean()
average_heartbeat
```

```python
## Round the averages to 2 decimal places for readability
average_age = round(average_age, 2)
average_heartbeat = round(average_heartbeat, 2)
```

```python
## Display the final calculated averages
print("Average Age:", average_age)
print("Average Heartbeat:", average_heartbeat)
```
